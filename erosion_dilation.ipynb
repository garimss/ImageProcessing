```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
import matplotlib.image as mpimg
from pylab import *
```

```python
img = cv2.imread('car1.jpg', 0)
```

```python
plt.imshow(img, cmap = 'gray', interpolation = 'bicubic')
```

```python
plt.show()
```

```python
kernel = np.ones((5,5), np.uint8)
```

```python
img_erosion = cv2.erode(img, kernel, iterations=1)
plt.imshow(img_erosion, cmap = 'gray', interpolation = 'bicubic')
plt.show()
```

```python
img_dilation = cv2.dilate(img, kernel, iterations=1)
plt.imshow(img_dilation, cmap = 'gray', interpolation = 'bicubic')
plt.show()
```

```python
fig, a = subplots(1,3, figsize=(20, 20))
a[0].imshow(img, cmap = 'gray', interpolation = 'bicubic')
a[0].axis('on')
a[0].set_title('Original', fontsize=20)
a[1].imshow(img_erosion,  cmap = 'gray', interpolation = 'bicubic')
a[1].axis('on')
a[1].set_title('Erosion', fontsize=20)
a[2].imshow(img_dilation,  cmap = 'gray', interpolation = 'bicubic')
a[2].axis('on')
a[2].set_title('Dilation', fontsize=20)
```
